<span style="font-size:30px; color:black">Customer Count for Sizes</span>

<span style="font-size:30px; color:black">Quantity of Sizes</span>

```python
Size_Qty = df.groupby(['Size'],as_index = False)['Qty'].sum().sort_values(by='Qty', ascending = False)
sns.barplot(x = 'Size', y = 'Qty', data = Size_Qty)
```

<b><H4>Note: </H4>Under to the above graph, Most of the Qty buy M size and least Qty buy 4XL size </b>

<span style="font-size:30px; color:black">Customer Categories</span>

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
pt.figure(figsize=(10,5))
pt.hist(column_data, bins=20, edgecolor = 'Black')
pt.xticks(rotation = 90)
pt.show()
```

<b><H4>Note: </H4>Under to the above graph, Most of the customers bought T-shirts. </b>

<span style="font-size:30px; color:black">B2B or Retailer</span>

```python
##Prepare data for pie chart
B2B_check = df['B2B'].value_counts()

pt.pie(B2B_check, labels=B2B_check, autopct = '%1.1f%%')
pt.show()
```

```python
##Prepare data for pie chart
B2B_check = df['B2B'].value_counts()

pt.pie(B2B_check, labels = B2B_check.index, autopct ='%1.1f%%')
pt.axis('equal')
pt.show()
```

<b><H4>Note: </H4>Under to the above graph, 99.2% buyers are retailers while 0.8% are B2B buyers. </b>

<span style="font-size:30px; color:black">Categories Over Sizes</span>

```python
##Prepare data for Scatter Plot
x_label = df['Category']
y_label = df['Size']

pt.scatter(x_label,y_label)
pt.xlabel('Category')
pt.ylabel('Size')
pt.title('Scatter Plot')
pt.show()
```

<b><H4>Note: </H4>Under to the above graph, T-shirts available the highest sizes. </b>

```python
pt.figure(figsize=(12,6))
sns.countplot(x= 'ship-state', data=df)
pt.xlabel('ship-state')
pt.ylabel('count')
pt.title('Distibution of States')
pt.xticks(rotation = 90)
pt.show()
```

```python
##Top 10 states
top_10_states = df['ship-state'].value_counts().head(10)
pt.figure(figsize=(10,5))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)],x = 'ship-state')
pt.xlabel('ship-state')
pt.ylabel('count')
pt.xticks(rotation = 90)
pt.title('Distribution of States')
pt.show()
```

<b><H4>Note: </H4>Under to the above graph, Most of buyers are in Maharasthra </b>

<b><H2>Conclusion: </H2>When considering about these details we can see that the business has a significant customer base in Mahasthra state and mainly serves retailers through Amazon.Compairing to others the highest demand is for T-shirts with M sizes among buyers. </b>
